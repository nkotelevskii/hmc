```python
def plot_hist_1(dataset_name, d1, d2, d3, d4, d5, legend=False):
    labels = ['Recall@5', 'Recall@10', 'nDCG@100']

    x = np.arange(len(labels))  # the label locations
    width = 0.2  # the width of the bars

    inc = width / 2
    fig, ax = plt.subplots(figsize=(8, 7), dpi=300)
    r1 = ax.bar(x - 2 * inc, d1, width / 2, label='GlbAvg', color='tab:purple')
    r2 = ax.bar(x - 1 * inc, d2, width / 2, label='BPR')
    r3 = ax.bar(x, d3, width / 2, label='WRMF', color='y')    
    r4 = ax.bar(x + inc, d4, width / 2, label='MultiVAE', color='g')
    r5 = ax.bar(x + 2 * inc, d5, width / 2, label='MetVAE', color='r')

    ax.set_title(dataset_name)
    if dataset_name == 'Foursquare':
        ax.set_ylim((0, 0.17))
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
#     ax.grid()
    if legend:
        ax.legend(loc=0, prop={'size': 16})
    fig.tight_layout()
#     plt.show()
    plt.savefig('./' + dataset_name + '_1' + '.pdf', format='pdf')
```

```python
plot_hist_1('Foursquare', fs_glbavg, fs_bpr, fs_wrmf, fs_multivae, fs_metvae, True)
```

```python
plot_hist_1('Gowalla', go_glbavg, go_bpr, go_wrmf, go_multivae, go_metvae)
```

```python
plot_hist_1('MovieLens20M', ml_glbavg, ml_bpr, ml_wrmf, ml_multivae, ml_metvae)
```

```python
def plot_hist_2(dataset_name, d1, d2, d3, d4, legend=False):
    labels = ['Recall@5', 'Recall@10', 'nDCG@100']

    x = np.arange(len(labels))  # the label locations
    width = 0.2  # the width of the bars

    inc = width / 2
    fig, ax = plt.subplots(figsize=(8, 7), dpi=300)
    r1 = ax.bar(x - inc, d1, width / 2, label='MultiVAE', color='tab:purple')
    r2 = ax.bar(x, d2, width / 2, label=r'$\beta$-MultiVAE')
    r3 = ax.bar(x + inc, d3, width / 2, label='MetVAE', color='y')    
    r4 = ax.bar(x + 2 * inc, d4, width / 2, label=r'$\beta$-MetVAE', color='g')

    ax.set_title(dataset_name)
    if dataset_name == 'Foursquare':
        ax.set_ylim((0.04, 0.14))
    elif dataset_name == 'Gowalla':
        ax.set_ylim((0.08, 0.17))
    elif dataset_name == 'MovieLens20M':
        ax.set_ylim((0.25, 0.45))
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
#     ax.grid()
    if legend:
        ax.legend(loc=0, prop={'size': 16})
    fig.tight_layout()
#     plt.show()
    plt.savefig('./' + dataset_name + '_2' + '.pdf', format='pdf')
```

```python
plot_hist_2('Foursquare', fs_multivae, fs_bmultivae, fs_metvae, fs_bmetvae, True)
```

```python
plot_hist_2('Gowalla', go_multivae, go_bmultivae, go_metvae, go_bmetvae)
```

```python
plot_hist_2('MovieLens20M', ml_multivae, ml_bmultivae, ml_metvae, ml_bmetvae)
```
